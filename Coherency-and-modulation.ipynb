# Coherency

Understanding coherency in an intuitive manner with modifyiable examples. Coherency being a complex measure that tells us information about two signals. 

Signal $x(t)$ and $y(t)$ in the time domain. 

```python
# (*) To communicate with Plotly's server, sign in with credentials file
import plotly.plotly as py
# We plot both cross-spectra
import plotly.graph_objs as go
# (*) Useful Python/Plotly tools
from plotly.graph_objs import Line  # (*) import Line 

import plotly.tools as tls

import numpy as np  # (*) numpy for math functions and arrays

import json

import plotly

from plotly.graph_objs import *

from IPython.html import widgets 
from IPython.display import display, clear_output

from scipy.fftpack import fft
from scipy.signal import hanning, fftconvolve

from plotly.widgets import GraphWidget
```

```python
time = 6 # in seconds: time of recording per trial
Fs = 1000 # in Hertz: sampling frequency
T = 500 # in ms: time of an epoch

# We can already calculate the interrelationships between time and frequency domains
freq_res = 1/(T/1000) # frequency resolution
N = Fs*(T/1000) # number of samples in an epoch
Fmax = (N/2)*freq_res # maximal frequency or nyquist frequency
W = np.linspace(0,Fmax,T)*freq_res # here we store the frequency indices

# For the statistics
num_trials = 1000
```

```python
def sources(phase_val, noise_val, freq1_val, freq2_val):
    noise_coef = noise_val # noise coefficient
    t = np.linspace(0,time*Fs,time*Fs,endpoint=True) # time points in miliseconds, 10 seconds at 1000 Hz
    freq1 = freq1_val # Hz, first frequency
    freq2 = freq2_val # Hz, second frequency 

    phase_shift = phase_val*2*np.pi # in radians
    s1t = np.sin(2*np.pi*t*freq1/Fs) # let's say it is between both sensors, so no lag
    s2_1t = np.sin(2*np.pi*(t*freq2/Fs)) # and let's say signal 2 is closer to sensor 1
    # so it won't be shifted for sensor one, but it will be phase shifted for sensor 2
    s2_2t = np.sin(2*np.pi*(t*freq2/Fs)+ phase_shift)
    
    xt = fftconvolve(1.1*s1t,0.5*s2_1t,'same') + noise_coef*np.random.normal(0,1,time*Fs)
    yt = fftconvolve(0.1*s1t,0.7*s2_2t,'same') + noise_coef*np.random.normal(0,1,time*Fs)
    xt_old = s1t + 1.5*s2_1t + noise_coef*np.random.normal(0,1,time*Fs)
    yt_old = 0.5*s1t + .3*s2_2t + noise_coef*np.random.normal(0,1,time*Fs)
    return t,xt,yt
```

```python
def coherency(time,f,xt,yt,Fs=1000,T=500):
    #Making a matrix of the epochs of x(t)
    X = np.reshape(xt,(time*Fs/T,T))
    Y = np.reshape(yt,(time*Fs/T,T))

    # And we check the size of our epochs matrix
    np.shape(X)

    H_window = hanning(T)

    Xf = np.zeros((time*Fs/T,T),dtype="complex_")
    Yf = np.zeros((time*Fs/T,T),dtype="complex_")

    for i in range(np.int(time*Fs/T)):
        Xf[i,:]=fft(X[i,:]*H_window)
        dataA = Scatter(
            x=W, 
            y=np.abs(fft(X[i,:]*H_window)),
            line=Line(
                color='red'   # set line color to black
            )
        ) 
    
    for i in range(np.int(time*Fs/T)):
        Yf[i,:]=fft(Y[i,:]*H_window)
        dataB = Scatter(
            x=W, 
            y=np.abs(fft(Y[i,:]*H_window)),
            line=Line(
                color='blue'   # set line color to black
            )
        )

    # We saved the fourier transforms of the signals in Xf and Yf

    # We approximate the expectation value with the mean over all the epochs
    # We will take a narrow band to include +-2 Hz around f
    
    # First we convert f to columns according to the resolution of our fft
    w = np.int(f/freq_res)
    Sxy_f1=np.mean(np.mean(Xf[:,w-1:w+1])*np.conj(np.mean(Yf[:,w-1:w+1])))
    
    # We reuse the code for S_xy we just use xx or yy. 
    Sxx_f1=np.mean(np.mean(Xf[:,w-1:w+1])*np.conj(np.mean(Xf[:,w-1:w+1])))
    Syy_f1=np.mean(np.mean(Yf[:,w-1:w+1])*np.conj(np.mean(Yf[:,w-1:w+1])))

    # We now construct the denominator of the coherency
    deno_f1 = np.sqrt(np.real(Sxx_f1*Syy_f1))
    
    # We finally get the coherency
    Cxy_f1=Sxy_f1/deno_f1
    
    return Cxy_f1
```

```python
def many_trials(phase, noise, freq1, freq2, time, f, num_trials=100):
    Cohs = list()
    for i in range(num_trials):
        t,xt,yt = sources(phase, noise, freq1, freq2)
        Cohs.append(coherency(time,f,xt,yt))
    return Cohs
```

```python
g = GraphWidget('https://plot.ly/~chris/4103')
g2 = GraphWidget('https://plot.ly/~chris/4103')
g4 = GraphWidget('https://plot.ly/~chris/4103')

slider_phase = widgets.IntSliderWidget()
slider_phase.min = 0
slider_phase.max=99
slider_phase.value=3
slider_phase.description = 'Percent of phase shift'

slider_noise = widgets.IntSliderWidget()
slider_noise.min = 0
slider_noise.max=100
slider_noise.value=1
slider_noise.description = 'Percent of white noise amplitude'

slider_freq1 = widgets.IntSliderWidget()
slider_freq1.min = 1
slider_freq1.max=80
slider_freq1.value=10
slider_freq1.description = 'Frequency of signal s1 (Hz)'

slider_freq2 = widgets.IntSliderWidget()
slider_freq2.min = 1
slider_freq2.max=80
slider_freq2.value=37
slider_freq2.description = 'Frequency of signal s2 (Hz)'

slider_f = widgets.IntSliderWidget()
slider_f.min=0
slider_f.max=80
slider_f.value=10
slider_f.description='Frequency on which to compute the coherency (Hz)'


def regraph(_):
    tt,xt,yt = sources(slider_phase.value/100, slider_noise.value/100,slider_freq1.value,slider_freq2.value)
    Cxy_f = coherency(time,slider_f.value,xt,yt)
    cohe = many_trials(slider_phase.value/100, slider_noise.value/100,slider_freq1.value,slider_freq2.value,time,slider_f.value, num_trials)
    # Since all coherency values lay on the unit circle, we actually only care about their angle. 
    # To extract the angle we use the arctan2() function. In this way we go from a complex number 
    # to a real one. The arctan2() function returns the “four quadrant” arctan of the angle formed 
    # by (x, y) and the positive x-axis. It returns values in radians in the (-pi, pi) range.

    angles = np.arctan2(np.imag(cohe),np.real(cohe))
    bb = 80
    counts,bins = np.histogram(angles+np.pi, bins=bb, range=[0,2*np.pi*(1 + 1/bb)], normed=True, weights=None, density=None)

    g.restyle({'x': [[]], 'y': [[]]})
    g.add_traces(dict(x=tt[0:1000], y=xt[0:1000],name = 'Source 1: x(t)'))                     
    g.add_traces(dict(x=tt[0:1000], y=yt[0:1000],name = 'Source 2: y(t)'))
    g.relayout({'xaxis.title': 'Time [ms]'})
    
    g2.restyle({'x': [[]], 'y': [[]]})
    
    trace1 = go.Scatter(
        x = [np.real(Cxy_f)],
        y = [np.imag(Cxy_f)],
        name='Coherency_xy(f)'
    )
    
    trace2 = go.Scatter(
        x = [0,np.real(Cxy_f)],
        y = [0,np.imag(Cxy_f)],
        mode = 'lines'
    )

    g2.add_traces(trace1)
    g2.add_traces(trace2)
    g2.relayout({'xaxis.range': [-1.2, 1.2]})
    g2.relayout({'yaxis.range': [-1.2, 1.2]})
    g2.relayout({
    'width':500,
    'height':500,
    'shapes': [
        {
            'type': 'circle',
            'xref': 'x',
            'yref': 'y',
            'x0': -1,
            'y0': -1,
            'x1': 1,
            'y1': 1,
            'opacity': 0.2,
            'fillcolor': 'white',
            'line': {
                'color': 'black',
            }
        }]})
    
    trace_hist = go.Bar(
    y = counts,
    x = bins,
    name = 'Coherency_xy(f)'
    )
    g2.relayout({'title': 'Coherency between x(t) and y(t) for one trial'})
    
    g4.restyle({'x': [[]], 'y': [[]]})
    g4.add_traces(trace_hist)
    g4.relayout({'xaxis.title': 'Angle of coherency (0 to 2pi)'})
    g4.relayout({'title': 'Histogram of coherency angles from 1000 simulations'})
    #g4.relayout({'xaxis.tick0': 0,'xaxis.dtick': 0.5*np.pi, 'xaxis.ticksuffix': '$\pi$',
    #             'xaxis.showticksuffix': 'last', 'xaxis.ticks': ['0','.5','1','1.5','2']})

    
slider_phase.on_trait_change(regraph, 'value')
slider_noise.on_trait_change(regraph, 'value')
slider_freq1.on_trait_change(regraph, 'value')
slider_freq2.on_trait_change(regraph, 'value')
slider_f.on_trait_change(regraph, 'value')    
    
display(slider_phase)
display(slider_noise)
display(slider_freq1)
display(slider_freq2)
display(slider_f)
display(g)    
display(g2)
display(g4)
```
